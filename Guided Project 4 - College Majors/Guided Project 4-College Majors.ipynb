```python
raw_data_count = 173
recent_grads = recent_grads.dropna()
```

```python
(recent_grads.plot
        (x='Sample_size', 
         y='Median', 
         kind='scatter', 
         title='Median salary vs. Sample_size', 
         figsize=(5,5)
         )
)
```

No relationship between median annual income and sample size (There shouldn't be a relationship).  Small sample size for most of the majors.

```python
(recent_grads.plot
        (x='Sample_size', 
         y='Unemployment_rate', 
         kind='scatter', 
         title='Unemployment Rate vs. Sample Size', 
         figsize=(5,5)
         )
)
```

No relationship between unemployment rate and sample size (There shouldn't be a relationship).  Small sample size for most of the majors.

```python
(recent_grads.plot
        (x='Full_time', 
         y='Median', 
         kind='scatter', 
         title='Full time employment vs. Median salary', 
         figsize=(7,5)
         )
)
```

No relationship between median salary and no. of grads in full time employment. 

```python
(recent_grads.plot
         (x='Median', 
          y='ShareWomen', 
          kind='scatter', 
          title='Median Salary vs. Share of Women', 
          figsize=(7,5)
          )
)            
```

There is a possible (very weak) relationship between share of women (expressed as a probability between 0 and 1) and median salary.  As the median salary increases the share of women decreases.

```python
(recent_grads.plot
        (x='Men', 
         y='Median', 
         kind='scatter', 
         title='Median salary vs. # of Men', 
         figsize=(7,5)
         )
)
```

No relationship between median salary and number of men.

```python
(recent_grads.plot
        (x='Women', 
         y='Median', 
         kind='scatter', 
         title='Median salary vs. # of Women', 
         figsize=(7,5)
         )
)
```

No relationship between median salary and number of women.

The following questions were asked for this project based on these scatterplots (*answers in italics*):

- Do students in more popular majors make more money? *The scatter plots of median vs. men,women do not show a relationship*

- Do students that majored in subjects that were majority female make more money? *The scatter plot of median salary vs. share of women suggest a weak relationship that majority women make **less** money*

- Is there any link between the number of full-time employees and median salary? *The scatter plots of median vs. full time do not show a relationship*
