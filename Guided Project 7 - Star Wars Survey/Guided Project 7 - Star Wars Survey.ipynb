```python
#rename ranking columns
SWS.rename(columns =
            {
            'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "Rank_film_1", 
            'Unnamed: 10': "Rank_film_2",'Unnamed: 11': "Rank_film_3", 'Unnamed: 12': "Rank_film_4", 
            'Unnamed: 13': "Rank_film_5", 'Unnamed: 14':"Rank_film_6" 
            }, inplace = True
           )
```

```python
SWS.iloc[:,9:15].head(5)
```

```python
fillna_cols = {"Rank_film_1": 0,"Rank_film_2" : 0, "Rank_film_3" : 0, "Rank_film_4" : 0, "Rank_film_5" : 0,"Rank_film_6": 0}
rank_cols = ["Rank_film_1","Rank_film_2", "Rank_film_3", "Rank_film_4", "Rank_film_5","Rank_film_6"]
SWS.fillna(value = fillna_cols)
```

```python
SWS.iloc[:,9:15].head(2)
```

```python
#convert rank columns to floats and check
SWS[SWS.columns[9:15]] = SWS[SWS.columns[9:15]].astype(float)
SWS.dtypes
```

```python
rank_avgs = SWS[rank_cols].mean(axis = 0)
rank_avgs
```

```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
ax_rank = rank_avgs.plot.bar(color = 'gray')

ax_rank.set_ylabel('Average Ranking', size = 14, color = 'g')
ax_rank.set_xlabel('Film (1 is Phantom Menace)', size = 14, color = 'g')
ax_rank.set_title('Respondent Ranking of Star Wars Films (1-6, 1 is best)', size = 18, color = 'g')

plt.show()
```

CONCLUSION

Films 4 - 6 are more favorable with respondents than 1 -3, with 5 ("The Empire Strikes Back") being the fan favorite for all 6 films.

From researching on Googles Films 1 - 3 probably suffered from unrealistic fan expectations, being prequels to 4 - 6. THe dialog was poorly written, and Anakin/Padme love story was flat and Jar Jar Binks was a terrible character.

Film 5 is the darkest, introduces the popular character Yoda, and provides the big reveal about Lukes lineage.  The line "I am your father" is still used in pop culture today.

```python
seen_cols = ["Seen_1","Seen_2", "Seen_3", "Seen_4", "Seen_5","Seen_6"]
seen_counts = SWS[seen_cols].sum(axis = 0)
seen_counts
```

```python
fig = plt.figure(figsize=(15,6))

ax_rank = fig.add_subplot(1,2,1)
ax_rank = rank_avgs.plot.bar(color = 'gray')
ax_rank.set_ylabel('Average Ranking', size = 14, color = 'g')
ax_rank.set_xlabel('Film (1 is Phantom Menace)', size = 14, color = 'g')
ax_rank.set_title('Respondent Ranking of Star Wars Films (1-6, 1 is best)', size = 16, color = 'g')

ax_seen = fig.add_subplot(1,2,2)
ax_seen = seen_counts.plot.bar(color = 'gray')
ax_seen.set_ylabel('Counts', size = 14, color = 'b')
ax_seen.set_xlabel('Film (1 is Phantom Menace)', size = 14, color = 'b')
ax_seen.set_title('# of Respondents Seen Star Wars Film', size = 16, color = 'b')

plt.show()
```

CONCLUSION

The highest ranked film (5) was also one of the most frequently watched films.  Disapointment (low ranking) among movie goers increased from films 1 - 3, accompanied by a decrease in attendance from films 1 - 3.
